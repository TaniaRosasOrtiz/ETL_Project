```python
import pandas as pd
```

```python
data_path= 'Resources/Petroleum_Well.csv'
wells_df=pd.read_csv(data_path)
```

```python
#Converting csv file into Data Frame
wells_df=pd.DataFrame(wells_df).copy()
wells_df.head()
```

```python
#geting unique values fo well_mode and creating a Data Frame with them
well_mode= wells_df['WELL_MODE'].unique()
cat_well_mode=pd.DataFrame(well_mode)
cat_well_mode.columns=['CAT well mode']
```

```python
cat_well_mode.head()
```

```python
wells_df['OPERATOR']=wells_df['OPERATOR'].str.upper()
```

```python
operator= wells_df['OPERATOR'].unique()
cat_operator=pd.DataFrame(operator)
cat_operator.columns=['Operator']
```

```python
#sorting values 
cat_operator = cat_operator.sort_values(['Operator'])
cat_operator.reset_index(drop=True, inplace=True)
```

```python
#exporting dfs to csv files
cat_well_mode.to_csv(r'Resources/cat_well_mode.csv', index=False)
cat_operator.to_csv(r'Resources/cat_operator.csv', index=False)
```
