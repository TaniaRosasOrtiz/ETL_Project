## LOGS

```python
import pandas as pd

# Import SQL Alchemy
from sqlalchemy import create_engine

# Import and establish Base for which classes will be constructed 
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()

# Import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float
```

```python
data_path= 'Resources/OGSR-Library-logs-2013-02-13.csv'
logs_df=pd.read_csv(data_path)
```

```python
logs_df.reset_index(drop=False, inplace = True)
logs_df.head()
```

```python
logs_df['type']=logs_df['type'].str.upper()
log_type = logs_df['type'].unique()
cat_log_type=pd.DataFrame(log_type)
cat_log_type.reset_index(drop=False, inplace = True)
cat_log_type.columns=['log_type_id','log_type']
# cat_log_type.set_index('log_type_id', inplace = True)
cat_log_type.head()
```

```python
cat_log_type.to_csv(r'Resources/cat_log_type.csv')
```

```python
log_hdr = logs_df.loc[:, ["index", "name", "type","licence", "total depth date", "latitude", "longitude"]]
log_hdr = pd.merge(left=logs_df, right=cat_log_type, left_on='type', right_on='log_type', how='left')
log_hdr["utm"] = 17
logs = pd.DataFrame(log_hdr[["index", "name", "log_type_id","licence", "total depth date", "longitude", "latitude", "utm"]])
logs.rename(columns = {'index':'log_id', 'name':'log_name',
                                 'licence':'license', 'total depth date':'total_depth_date',
                                 'longitude':'x', 'latitude':'y'
                                }, inplace = True)
logs.head(15)
```

```python
cat_log_type.set_index('log_type_id', inplace = True)
logs.set_index('log_id', inplace = True)
```

```python
# Create a connection to a SQLite database
connection_string = "postgres:pg@localhost:5432/ETL_PROJECT"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
cat_log_type.to_sql(name='cat_log_type', con=engine, if_exists='append', index=True)
```

```python
logs.to_sql(name='log_hdr', con=engine, if_exists='append', index=True)
```
